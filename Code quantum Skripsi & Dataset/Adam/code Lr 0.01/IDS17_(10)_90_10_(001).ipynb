```python
np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)
np.random.seed(1967)
x, y = zip(*np.random.permutation(list(zip(x, y))))

pca = sklearn.decomposition.PCA(n_components=n_features)
pca.fit(x)
x = pca.transform(x)
```

```python
x.shape
```

```python
x_min = np.min(x, axis=0)
x_max = np.max(x, axis=0)

x = 2 * np.pi * (x - x_min) / (x_max - x_min) - np.pi
```

```python
split = int(n_samples - (n_samples*0.1))
print("Data train :", split)
print("Data test :", n_samples-split)

x_train = x[:split]
x_test = x[split:]
y_train = y[:split]
y_test = y[split:]
```

```python
n_wires = 4

dev0 = qml.device("rigetti.qvm", device="4q-qvm")
dev1 = qml.device("qiskit.aer", wires=4)
devs = [dev0, dev1]
```

```python
def circuit0(params, x=None):
    for i in range(n_wires):
        qml.RX(x[i % n_features], wires=i)
        qml.Rot(*params[1, 0, i], wires=i)

    qml.CZ(wires=[1, 0])
    qml.CZ(wires=[1, 2])
    qml.CZ(wires=[3, 0])

    for i in range(n_wires):
        qml.Rot(*params[1, 1, i], wires=i)
    return qml.expval(qml.PauliZ(0)), qml.expval(qml.PauliZ(1))


def circuit1(params, x=None):
    for i in range(n_wires):
        qml.RX(x[i % n_features], wires=i)
        qml.Rot(*params[0, 0, i], wires=i)

    qml.CZ(wires=[0, 1])
    qml.CZ(wires=[1, 2])
    qml.CZ(wires=[1, 3])

    for i in range(n_wires):
        qml.Rot(*params[0, 1, i], wires=i)
    return qml.expval(qml.PauliZ(0)), qml.expval(qml.PauliZ(1))
```

```python
qnodes = [
    qml.QNode(circuit0, dev0, interface="torch"),
    qml.QNode(circuit1, dev1, interface="torch"),
]
```

```python
import tensorflow as tf
```

```python
n_classes = 2 #banyaknya label
n_layers = 2
# indeks untuk model
params = torch.tensor(np.random.random((2, n_layers, n_wires, 3)), requires_grad=True)
iterasi = 1

def softmax_ensemble(params, x_point=None):
    results = [q(params, x=x_point) for q in qnodes]
    results_tensor = torch.tensor(results)
    softmax = torch.nn.functional.softmax(results_tensor, dim=1)
    choice = torch.where(softmax == torch.max(softmax))[0][0]
    return softmax[choice]

def cost(params, y_point, x_point=None):
    return torch.sum(torch.abs(softmax_ensemble(params, x_point=x_point) - y_point))


y_soft = torch.tensor(tf.one_hot(y_train, n_classes).numpy(), requires_grad=True)

opt = torch.optim.Adam([params], lr = 0.01) #learning rate

for x_point, y_point in zip(x_train, y_soft):
    opt.zero_grad()
    c = cost(params, y_point=y_point, x_point=x_point)
    c.backward()
    opt.step()
    if iterasi % 10 == 0 and iterasi > 0:
        print ("Iterasi ke : ", iterasi)
    iterasi += 1
    
```

```python
def decision(softmax):
    return int(torch.argmax(softmax))


def predict_point(params, x_point=None, parallel=True):
    results = [q(params, x=x_point)for q in qnodes]
    results_tensor = torch.tensor(results)
    softmax = torch.nn.functional.softmax(results_tensor, dim=1)
    choice = torch.where(softmax == torch.max(softmax))[0][0]
    chosen_softmax = softmax[choice]
    return decision(chosen_softmax), decision(softmax[0]), decision(softmax[1]), int(choice)
```

```python
def predict(params, x=None, parallel=True):
    predictions_ensemble = []
    predictions_0 = []
    predictions_1 = []
    choices = []

    for i, x_point in enumerate(x):
        if i % 10 == 0 and i > 0:
            print("Completed up to iteration {}".format(i))
        results = predict_point(params, x_point=x_point, parallel=parallel)
        predictions_ensemble.append(results[0])
        predictions_0.append(results[1])
        predictions_1.append(results[2])
        choices.append(results[3])

    return predictions_ensemble, predictions_0, predictions_1, choices
```

```python
print("Predicting on training dataset")
p_train, p_train_0, p_train_1, choices_train = predict(params, x=x_train)
print("Predicting on test dataset")
p_test, p_test_0, p_test_1, choices_test = predict(params, x=x_test)
```

```python
def accuracy(predictions, actuals):
    count = 0

    for i in range(len(predictions)):
        if predictions[i] == actuals[i]:
            count += 1

    accuracy = count / (len(predictions))
    return accuracy
```

```python
print("Training accuracy (ensemble): {}".format(accuracy(p_train, y_train)))
print("Training accuracy (QPU0):  {}".format(accuracy(p_train_0, y_train)))
print("Training accuracy (QPU1):  {}".format(accuracy(p_train_1, y_train)))
```

```python
print("Test accuracy (ensemble): {}".format(accuracy(p_test, y_test)))
print("Test accuracy (QPU0):  {}".format(accuracy(p_test_0, y_test)))
print("Test accuracy (QPU1):  {}".format(accuracy(p_test_1, y_test)))
```

```python
def conf(predictions, actuals):
    tp = 0
    tn = 0
    fp = 0
    fn = 0

    for i in range(len(predictions)):
        if predictions[i] == 0:
            if predictions[i] == actuals[i]:
                tn += 1
            else:
                fn += 1
        else:
            if predictions[i] == actuals[i]:
                tp += 1
            else:
                fp += 1 
            
    return tp, tn, fp, fn
```

```python
train_tp, train_tn, train_fp, train_fn = conf(p_train, y_train)
print("Train True Positive :", train_tp)
print("Train True Negative :", train_tn)
print("Train False Positive :", train_fp)
print("Train False Negative :", train_fn)
```

```python
test_tp, test_tn, test_fp, test_fn = conf(p_test, y_test)
print("Test True Positive :", test_tp)
print("Test True Negative :", test_tn)
print("Test False Positive :", test_fp)
print("Test False Negative :", test_fn)
```
